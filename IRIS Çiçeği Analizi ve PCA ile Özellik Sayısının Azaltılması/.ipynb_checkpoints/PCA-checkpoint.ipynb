```python
# PCA -> çok boyutlu(feature) veri setlerini, veri setinin anlamını kaybetmeden daha az boyutlu hale getirme işlemi denebilir kısaca
# Bu işlem sırasında varyans %90'dan fazla bir oranda korunmaya çalışılır. yani min kayıp.
```

```python
# Veri setinde bulunan 4 adet feature'u 2 boyuta indirgeyeceğiz.

# Iris veri seti 3 Iris bitki türüne (Iris setosa, Iris virginica ve Iris versicolor) ait, her bir türden 50 örnek olmak üzere toplam 150 örnek sayisina sahip bir veri setidir. 
# Her bir örnek için 4 özellik tanimlanmistir: taç yaprak uzunlugu, taç yaprak genisligi, çanak yaprak genisligi, çanak yaprak uzunluğu('sepal length','sepal width','petal length','petal width'). 

# Veri setimizde, her bir bitki örnegi ayri bir gözlemi (örnegi) ifade ederken; 
# bitki tür ismi bagimli(dependent) degisken, bitkilerin ölçülen 4 temel özelligi ise bagimsiz(independent) degiskenleri ifade eder.
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("pca_iris.data", names=["sepal length","sepal width","petal length","petal width","target"])
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.tail()
```

```python
# x -> [features]
# y -> target
```

```python
features = ["sepal length","sepal width","petal length","petal width"]
x = data[features]
```

```python
y = data["target"]
```

#### Değerleri Scale etmemiz gerekiyor. Çünkü her bir feature çok farklı boyutlarda ve bunların yapay zeka tarafından eşit ağırlıklarda dengelenmesi gerekiyor. 

#### Bu amaçla standart scaler  kullanarak tüm verileri mean = 0 and variance = 1 olacak şekilde değiştiriyoruz.

PCA, veri setindeki bileşenleri oluştururken kovaryans matrisini kullanır. 
Kovaryans, değişkenlerin birbirleriyle nasıl değiştiğini ölçer. 
Eğer özelliklerin (değişkenlerin) ölçekleri farklıysa, bazı değişkenler gereğinden fazla etkili olabilir.

Örneğin, bir veri setinde:

Sepal Length = 4 - 7 cm aralığında,
Petal Width = 0.1 - 2.5 cm aralığında

Burada, Sepal Length daha büyük değerlere sahip olduğu için, PCA’nın oluşturduğu bileşenlerde daha fazla ağırlık kazanacaktır.
Bu durum, PCA’nın gerçek ilişkilere değil, ölçek farklılıklarına göre karar vermesine yol açar.

Standardizasyon, tüm değişkenlerin ortalamasını 0'a, standart sapmasını 1'e çekerek ölçek farklılıklarını ortadan kaldırır. Böylece:

Her özelliğin PCA analizi üzerindeki etkisi eşitlenir.
Değişkenler arasında adil karşılaştırma yapılır.
Kovaryans matrisindeki büyük değişim, yalnızca verideki ilişkileri yansıtır; ölçek farkından kaynaklanmaz.

```python
x_scaled = StandardScaler().fit_transform(x) 
```

```python
print("Scale edilmeden önce:\n", x)
print("------------------------------------------------------------")
print("Scale edildikten sonra:\n",x_scaled)
```

```python
# PCA: 4 boyuttan -> 2 boyuta

# Orijinal veriler 4 boyuta sahip: "sepal length","sepal width","petal length","petal width"
# Biz PCA yaparak bunları 2 boyuta indirgeyeceğiz ancak şunu belirtmeliyim ki PCA indirgeme işlemi sonucunda elde edeceğimiz 2 boyutun herhangi bir anlam ifade etmeyen başlıklara sahip olacak.. 
# Yani 4 feature'dan 2 tanesini basit bir şekilde atmak değil yapıla işlem..

pca = PCA(n_components=2)
pca_components = pca.fit_transform(x)
pca_data = pd.DataFrame(data = pca_components, columns=["pca_components_1", "pca_components_2"])
```

```python
pca_data
```

```python
# boyut küçükltme işleminden sonra target sütununu da ekleyip veri setimizi tekrardan bir bütün halde görelim.

final_data = pd.concat([pca_data, y], axis=1)
final_data.head()
```

```python
# şimdi görselleştirelim.. 

df_setosa = final_data[final_data.target == "Iris-setosa"]
df_virginica = final_data[final_data.target == "Iris-virginica"]
df_versicolor = final_data[final_data.target == "Iris-versicolor"]

plt.xlabel("pca_components_1")
plt.ylabel("pca_components_2")

plt.scatter(df_setosa["pca_components_1"], df_setosa["pca_components_2"], color="green")
plt.scatter(df_virginica["pca_components_1"], df_virginica["pca_components_2"], color="red")
plt.scatter(df_versicolor["pca_components_1"], df_versicolor["pca_components_2"], color="blue")

plt.show()
```
