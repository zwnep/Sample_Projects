```python
import numpy as np # linear algebra
import pandas as pd # data processing, csv file I/O etc.
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

from kmodes.kprototypes import KPrototypes  
from kmodes.kmodes import KModes
```

```python
data = pd.read_csv("segmentation_data.csv")
data
```

```python
# Görüldüğü gibi cinsiyet, eğitim gibi kategorik değerler var ancak gelir gibi de nümerik değerler bulunuyor.
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
# Nümerik değerlerin normalize etmemiz gerekiyor. 

data_temp = data[["Age","ID","Income"]] # normalizasyon öncesi orijinal değerleri tutuyoruz.. ne olur ne olmaz :)
#data_temp
```

```python

scaler = StandardScaler()

scaler.fit(data[["Age"]])
data["Age"] = scaler.transform(data[["Age"]])

scaler.fit(data[["Income"]])
data["Income"] = scaler.transform(data[["Income"]])

# ID sütunu gerekli oladığı içi scale edilmesine gerek yok..
```

```python
print(data.Age)
print(data.Income)
```

```python
data = data.drop(["ID"],axis=1)
data.head()
```

```python

# Modeli oluşturalım..

model = KPrototypes(n_clusters=10, verbose=2, max_iter=20)

clusters = model.fit_predict(data, categorical=[0, 1, 3, 5, 6])
```

```python
print(model.cluster_centroids_)

len(model.cluster_centroids_)
```

```python

cluster_dict = []
for c in clusters:
    cluster_dict.append(c)
```

```python

data["cluster"] = cluster_dict
```

```python
data[data['cluster']== 0].head(10)
```

```python

data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
data3 = data[data.cluster==2]
data4 = data[data.cluster==3]
data5 = data[data.cluster==4]
data6 = data[data.cluster==6]
data7 = data[data.cluster==7]
data8 = data[data.cluster==8]
data9 = data[data.cluster==9]
data10 = data[data.cluster==10]
```

```python
plt.figure(figsize=(15,15))
plt.xlabel('Age')
plt.ylabel('Income')

plt.scatter(data1.Age, data1['Income'],color='green', alpha = 0.4)
plt.scatter(data2.Age, data2['Income'],color='red', alpha = 0.4)
plt.scatter(data3.Age, data3['Income'],color='gray', alpha = 0.4)
plt.scatter(data4.Age, data4['Income'],color='orange', alpha = 0.4)
plt.scatter(data5.Age, data5['Income'],color='yellow', alpha = 0.4)
plt.scatter(data6.Age, data6['Income'],color='cyan', alpha = 0.4)
plt.scatter(data7.Age, data7['Income'],color='magenta', alpha = 0.4)
plt.scatter(data8.Age, data8['Income'],color='gray', alpha = 0.4)
plt.scatter(data9.Age, data9['Income'],color='purple', alpha = 0.4)
plt.scatter(data10.Age, data10['Income'],color='blue', alpha = 0.4)


plt.legend()
plt.show()
```

```python
# KModes algoritması, tamamen kategorik veriler için geliştirilmiş bir kümelere ayırma (clustering) yöntemidir. 
# Standart K-means algoritması, yalnızca sayısal (nümerik) verilerle çalışabilir. 
# Ancak birçok gerçek dünya problemlerinde, veriler kategorik (örneğin: renk, şehir, marka) olabilir. 
# KModes, bu tür verileri kümelere ayırmak için kullanılır.

# K-means, yalnızca sayısal verilerde etkilidir. Ancak birçok gerçek dünya veri kümesi kategorik veya karma verilere sahiptir.
# KModes ve KPrototypes, K-means algoritmasını genişleterek daha fazla veri türüyle çalışmamıza olanak tanır.

# https://pypi.org/project/kmodes/
```
