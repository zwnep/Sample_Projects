```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
columns = ['user_id', 'item_id', 'rating', 'timestamp'] # veri setini daha önce bastırdığımda sütun isimlerini olmadığını gördüm.

data_user = pd.read_csv("users.data", sep="\t", names=columns)
data_user.head()
```

```python
print("Kayıt sayısı:",len(data_user))
print("-------------------------------")
print(data_user.info())
print("-------------------------------")
print(data_user.describe())
```

```python
# Şimdi diğer dosyamızı yükleyelim..

data_movie = pd.read_csv("movie_id_titles.csv")
data_movie.head()
```

```python
print("Film sayısı:",len(data_movie))
print("-------------------------------")
print(data_movie.info())
print("-------------------------------")
print(data_movie.describe())
```

```python
data = pd.merge(data_user, data_movie, on="item_id")
data.head()
```

```python
data.tail()
```

```python
len(data)
```

```python
# pivot table oluşturucaz. excel'deki mantıkla aynı şekilde.

data_table = data.pivot_table(index="user_id", columns="title",values="rating")
data_table.head()
```

```python
# Bu yapıya göre her satır bir kullanıcı olacak şekilde (yani dataframe'imizin index'i user_id olacak)
# Sütunlarda film isimleri (yani title sütunu) olacak,
# tablo içinde de rating değerleri olacak şekilde bir dataframe oluşturuyoruz!
```

```python
type(data_table)
```

```python
# Amaç: Star Wars filmine benzer film önerileri yapmak.
```

```python
starwars_rating = data_table["Star Wars (1977)"]
starwars_rating.head()  
```

```python
similar_to_starwars = data_table.corrwith(starwars_rating)
similar_to_starwars  
```

```python
type(similar_to_starwars)
```

```python
# görüldüğü üzere similar_to_starwars'da boşluklar mevcut aynı zamanda bir serie 
# NaN değerlerini silip (eksik değerleir doldurabiliriz silmek yerine) df haline getirelim.

similar_starwars = pd.DataFrame(similar_to_starwars, columns=["Correlation"])
similar_starwars
```

```python
similar_starwars.sort_values("Correlation", ascending=False).head(10)
```

```python
# çok alakasız bir sıralama olduğu görülmektedir. neden peki??

data
```

```python
data.drop("timestamp", axis=1, inplace=True) # timestamp ihtiyaç yok.
data.head()
```

```python
# her filmin ortalama rating değerlerine bakalım..

rating = pd.DataFrame(data.groupby("title")["rating"].mean())
rating.head()
```

```python
# rating adında bir df oluşturuldu ancak rating sütununda ortalamalar hesaplanırken kaç oy aldığını bilmiyoruz.

rating["rating_count"] = pd.DataFrame(data.groupby("title")["rating"].count())
rating
```

```python
rating.sort_values("rating_count", ascending=False) # büyükten küçüğe sıraladık.
```

```python
# alakasız öneriler yapmasının sebebi 100 ün altında oylama yapılası yani az sayıda oy sayısının olması etkiledi.
```

```python
similar_starwars = similar_starwars.join(rating["rating_count"])
similar_starwars.head()
```

```python
# 100 den büyük oy sayısı olanları kullanarak sıralayalım.

similar_starwars[similar_starwars["rating_count"] > 100].sort_values("Correlation", ascending=False)
```
